```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import category_encoders as ce
import joblib
```

```python
encoder=ce.OneHotEncoder(cols=['gender','marital_status','education_level','employment_status','loan_purpose','grade_subgrade'],handle_unknown='return_nan',return_df=True,use_cat_names=True)
```

```python
ridge_model = joblib.load("ridge_model.joblib")
lasso_model = joblib.load("lasso_model.joblib")
DT_model = joblib.load("DT_model.joblib")
RF_model = joblib.load("RF_model.joblib")
XGB_model = joblib.load("XGB_model.joblib")
LGB_model = joblib.load("LGB_model.joblib")
CAT_model = joblib.load("CAT_model.joblib")
```

```python
df_test = pd.read_csv("test.csv")
df_test = df_test.drop(["id"], axis = 1)
df_test= encoder.fit_transform(df_test)
df_test = pd.get_dummies(data=df_test,drop_first=True)
df_test= (df_test - df_test.min())/(df_test.max() - df_test.min()) 
df_test = df_test.sort_index(axis=1)
df_test.head()
```

```python
predictions = ridge_model.predict(df_test)
np.savetxt('output_ridge.csv', predictions, delimiter=',')
df_new = pd.read_csv("test.csv")
```

```python
out = pd.read_csv("output_ridge.csv")
out["id"] = df_new["id"]
order = ["id","loan_paid_back"]
out = out[order]
out.to_csv("predictions_ridge.csv", index=False)
out.head()
```
