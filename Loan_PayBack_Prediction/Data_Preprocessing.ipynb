```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import category_encoders as ce
```

```python
df = pd.read_csv("kaggle_LoanPaybackModel/train.csv")
df.head()
```

```python
df.describe()
```

```python
df.corr(numeric_only=True)
```

```python
Y = df["loan_paid_back"]
X = df.drop(["id","loan_paid_back"], axis = 1)
```

```python
encoder=ce.OneHotEncoder(cols=['gender','marital_status','education_level','employment_status','loan_purpose','grade_subgrade'],handle_unknown='return_nan',return_df=True,use_cat_names=True)
```

```python
X= encoder.fit_transform(X)
X = pd.get_dummies(data=X,drop_first=True)
X.head()
```

```python
X = (X - X.min())/(X.max()-X.min())
```

```python
X.to_csv("X_Processed.csv", index=False)
Y.to_csv("Y_Processed.csv", index=False)
```
